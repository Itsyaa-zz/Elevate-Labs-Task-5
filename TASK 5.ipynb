```python

# 1. Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# 2. Load dataset
df = pd.read_csv('Sales Dataset.csv', encoding='ISO-8859-1')
```

```python
# 3. Basic info
print("Shape:", df.shape)
df.info()
display(df.describe())
display(df.head())
```

```python
# 4. Missing values
print("Missing values:")
print(df.isnull().sum())
```

```python
# 5. Univariate Analysis
# Histograms
df.hist(figsize=(16,12), bins=30)
plt.suptitle("Histograms of Numeric Features")
plt.show()
```

```python
# Boxplots
numeric_cols = df.select_dtypes(include=np.number).columns
plt.figure(figsize=(18,10))
for i, col in enumerate(numeric_cols):
    plt.subplot(3, 4, i+1)
    sns.boxplot(x=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# 6. Bivariate & Multivariate Analysis
# Correlation heatmap
numeric_df = df.select_dtypes(include=[np.number])

plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pairplot
selected = ['SALES', 'PRICEEACH', 'QUANTITYORDERED', 'MSRP']
sns.pairplot(df[selected])
plt.suptitle("Pairplot of Key Variables", y=1.02)
plt.show()
```

```python
# Scatter plot: Sales vs Price
plt.figure(figsize=(8,6))
sns.scatterplot(x='PRICEEACH', y='SALES', hue='DEALSIZE', data=df)
plt.title("Sales vs PriceEach by Deal Size")
plt.show()
```
